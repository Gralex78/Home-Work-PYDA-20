### Context  
Motivated by Gregory Smith's web scrape of VGChartz Video Games Sales, this data set simply extends the number of variables with another web scrape from Metacritic. Unfortunately, there are missing observations as Metacritic only covers a subset of the platforms. Also, a game may not have all the observations of the additional variables discussed below. Complete cases are ~ 6,900

### Content  
Alongside the fields:  

Name - Name of the game  
Platform - Console on which the game is running  
Year_of_Release - Year of the game released  
Genre - Game's category  
Publisher - Publisher   
NA_Sales - Game sales in North America (in millions of units)  
EU_Sales - Game sales in the European Union (in millions of units)  
JP_Sales - Game sales in Japan (in millions of units)  
Other_Sales - Game sales in the rest of the world, i.e. Africa, Asia excluding Japan, Australia, Europe excluding the E.U.  
Global_Sales - Total sales in the world (in millions of units)  
Critic_Score - Aggregate score compiled by Metacritic staff  
Critic_Count - The number of critics used in coming up with the Critic_Score    
User_Score - Score by Metacritic's subscribers  
User_Count - Number of users who gave the User_Score  
Developer - Party responsible for creating the game  
Rating - The ESRB ratings (E.g. Everyone, Teen, Adults Only..etc)

## 1. Очистка данных и заполнение пустых значений датафрейма

### 1.1 Типы данных  

- значения поля 'Year_of_Release' из float переведём в int (позже, после заполнения пропущенных значений)  
- значения поля 'User_Score' переведём из object (str) во float
- значения полей 'Critic_Count' и 'User_Count' из float в int (по-скольку они отражают количество человек, т.е. должно быть целое число)  
- значения полей 'Critic_Score' и 'User_Score' приведем к одному порядку

```python
# Переведем типы данных значений поля 'Year_of_Release' из float в int
df['Year_of_Release'] = df.Year_of_Release.astype('Int64')

# Переименуем поле 'Year_of_Release' в 'Year', для краткости отображения
df.rename(columns = {df.columns[2]: 'Year'}, inplace = True)

# Исправим ошибку в поле 'User_Score' и переведем типы данных значений из object во float
df['User_Score'].replace('tbd', np.NaN, inplace=True)
df['User_Score'] = df.User_Score.astype('Float64')

# Переведем типы данных значений полей 'User_Count' и 'Critic_Count' из float в int
df['User_Count'] = df.User_Count.astype('Int64')
df['Critic_Count'] = df.Critic_Count.astype('Int64')

# Приведем значения поля 'User_Score' к такому же порядку, как и у значений поля 'Critic_Score'
df['User_Score'] = df['User_Score'] * 10
```

Часть игр содержат в своих наименованиях ГОД, поэтому, как вариант, можно это использовать для заполнения пропусков поля 
Year_of_Release. Предполагаю, что эти игры выпускались либо в год указанный в наименовании, либо годом ранее или не позже, чем на следующий. Напишем функцию, которая будет с помощью регулярных выражений возвращать год из строчных значений поля 'Name'. В случае если год указан,  функция этими значениями заполнит пропуски в поле  'Year_of_Release'.

```python
import re
```

```python
# Функция возвращает год из названия игры, предварительно найдя его по заданному шаблону regex
#def get_year(string):
#    year = re.findall(r'\d{4}', string)   # шаблон
#    for string in df.Name:           # итерация по строковым значениям поля Name    
#        if len(year) != 0:      # функция фозвращает строковое значение в списке, если года не найдено возвращает пустой список
#            return int(year[0])  # 1-й элемент из списка, преобразованный в DataType: int

# добовляем столбец с полученными значениями (где год в поле "Наименование игры" отсутствует - заполнено: None)
#df['Year'] = df['Name'].apply(get_year).astype('Int64') 
```

```python
df['Year'].fillna(df['Name'].apply(lambda x: re.findall(r'\d{4}', x))\
                            .map(lambda x: int(x[0]) if len(x) != 0 else None).astype('Int64'), inplace=True)
```

```python
df.Year.isna().sum()
```

Далее для наглядности количества пропусков поля 'Year_of_Release', сгруппируем df по Наименованию игр (потому что игра, как правило, выпускается, для разных консолей, примерно в одно время, год-два) и агригируем значения атрибутов: 'Platform', 'Publisher', 'Developer', в разрезах которых, предположительно, можно заполнить пустые значения поля 'Year_of_Release'.    

```python
years_names = df.groupby('Name').agg(
    num_platf=('Platform', 'nunique'), # количество платформ
    count_platf=('Platform', 'count'), # количество изданий (т.е. для одной платформы могло издаваться несколько версий игры)
    num_publ=('Publisher', 'nunique'), # количество уникальных издателей
    count_publ=('Publisher', 'count'), # общее количество издателей
    size_publ=('Publisher', 'size'), # общее количество издателей с NaN
    num_devs=('Developer', 'nunique'), # количество уникальных разработчиков
    count_devs=('Developer', 'count'), # общее количество разработчиков
    size_devs=('Developer', 'size'), # общее количество разработчиков с NaN
    first_year=('Year', 'min'), # год первого издания (-й)
    last_year=('Year', 'max'), # последнего издания (-й)
    num_year=('Year', 'nunique'), # количество уникальных значений года (период времени релизов, в годах)
    size_year=('Year', 'size'), # size учитывает количество NaN, count - не учитывает
    count_year=('Year', 'count')).sort_values('num_year', ascending=False).reset_index()

# добавим столбцы с количеством пропусков полей:

# 'Year_of_Release'
years_names.loc[:, 'year_isna'] = years_names.size_year - years_names.count_year

# 'Publisher'
years_names.loc[:, 'publ_isna'] = years_names.size_publ - years_names.count_publ

# 'Developer'
years_names.loc[:, 'devs_isna'] = years_names.size_devs - years_names.count_devs

years_names = years_names[['Name', 'num_platf', 'count_platf', 'num_publ', 'count_publ', 'publ_isna', 'num_devs', 'count_devs',\
                           'devs_isna', 'first_year', 'last_year', 'num_year', 'count_year', 'year_isna']]

# отсортируем по количеству пропусков в поле 'Year_of_Release'
years_names.sort_values('year_isna', ascending=False).head(10)
```

```python
# years_names.loc[years_names.year_isna > 0].sort_values('year_isna', ascending=False).head(50)
```

Из полученных агригаций видна обратная связь количества пропусков в полях "Год издания" и "Издатель" и, в чуть меньшей степени, - "Год издания" и "Разработчик". Таким образом заполним, в разрезах "Наименование игры + Издатель" и "Наименование игры + Разработчик", поле "Год издания" самыми ранними значениями значениеми.

```python
df['Year'].fillna(df.groupby(['Name', 'Publisher'])['Year'].transform('min'), inplace=True)
df['Year'].fillna(df.groupby(['Name', 'Developer'])['Year'].transform('min'), inplace=True)
df['Year'].fillna(df.groupby(['Name'])['Year'].transform('min'), inplace=True)
df['Year'].fillna(df.groupby(['Platform', 'Publisher'])['Year'].transform('min'), inplace=True)
df['Year'].fillna(df.groupby(['Platform', 'Developer'])['Year'].transform('min'), inplace=True)
df['Year'].fillna(df.groupby(['Platform'])['Year'].transform('min'), inplace=True)
df.Year.isna().sum()
```
