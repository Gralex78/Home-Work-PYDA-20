```python
import pandas as pd
```

```python
ratings = pd.read_csv('ratings.csv')
ratings
```

```python
movies = pd.read_csv('movies.csv')
movies
```

```python
top_star = ratings['rating'].max()
top_star_rating = ratings.query('rating == 5.0'.format(top_star))
joined = top_star_rating.merge(movies, on='movieId', how='left')
print('НАИБОЛЬШЕЕ КОЛИЧЕСТВО ОЦЕНОК "5.0" БЫЛО ВЫСТАВЛЕНО ФИЛЬМУ: ', joined['title'].value_counts().head(1))
```

## Задание 2

По данным файла power.csv посчитайте суммарное потребление стран Прибалтики (Латвия, Литва и Эстония) категорий 4, 12 и 21 за период с 2005 по 2010 года. Не учитывайте в расчетах отрицательные значения quantity.

```python
power = pd.read_csv('power.csv')
power
```

```python
pribaltica_power = power[ ((power['country']=='Latvia') | (power['country']=='Lithuania') | (power['country']=='Estonia')) &\
                          ((power['category']==4) | (power['category']==12) | (power['category']==21)) &\
                          ((power['year']>=2005) & (power['year']<=2010)) & (power['quantity']>0) ]
print('Суммарное потребление стран Прибалтики категорий 4, 12 и 21 за период с 2005 по 2010 года составило: ',\
      pribaltica_power['quantity'].sum())
```
