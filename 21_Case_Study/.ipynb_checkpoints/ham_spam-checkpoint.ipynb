```python
import pandas as pd
import numpy as np
import re
```

```python
df = pd.read_html('https://github.com/obulygin/pyda_homeworks/blob/master/stat_case_study/spam.csv')[0]
df = df[['Category', 'Message']]
df
```

```python
messages = []
categories = []

for i in df:
    message = df.Message
    category = df.Category
    messages.append(message)
    categories.append(category)
    break    
```

```python
messages[0]
```

```python
# приведем все слова к нижнему регистру
texts_lower_case = [text.lower() for text in messages[0]]
# проверим
texts_lower_case[5569]
```

```python
# уберем все служебные символы
print(len(texts_lower_case[5569]))
only_words_text = [re.sub('[\W_]+',' ', text) for text in texts_lower_case]
# проверим
print(len(only_words_text[5569]))
only_words_text[5569]
```

```python
# разделим все на слова
texts_splitted = [text.split() for text in only_words_text]
# сколько слов?
len(texts_splitted[5569])
```

```python
texts_splitted[5569]
```

```python
from nltk.corpus import stopwords
```

```python
stopwords_set = set(stopwords.words('english'))
without_sw = [[word for word in text if word not in stopwords_set] for text in texts_splitted]
len(without_sw[5569])
```

```python
without_sw[5569]
```

```python
without_sw
```

```python
from nltk.stem import WordNetLemmatizer
```

```python
wordnet_lemmatizer = WordNetLemmatizer()
lemmatized_text = [[wordnet_lemmatizer.lemmatize(word) for word in text] for text in without_sw]
lemmatized_text[5569]
```

```python
lemmatized_text
```

```python
lemmatized_text_df = pd.DataFrame(lemmatized_text_joined_only_words)
lemmatized_text_df.columns = ['message']
lemmatized_text_df
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

tfidf = TfidfVectorizer()
tfidf_matrix = tfidf.fit_transform(lemmatized_text_df.message)
names = tfidf.get_feature_names()
tfidf_matrix = pd.DataFrame(tfidf_matrix.toarray(), columns=names)
tfidf_matrix
```
